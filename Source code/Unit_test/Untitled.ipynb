```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from IPython.display import set_matplotlib_formats

matplotlib.rc('font', family = 'Malgun Gothic')

matplotlib.rc('axes', unicode_minus=False)

set_matplotlib_formats('retina')
```

```python
Seoul_Parking = pd.read_csv("주차장정보.csv", encoding='euc-kr')

print(Seoul_Parking.shape)
Seoul_Parking.head()
```

```python
Gangnam_Parking = pd.read_csv("강남구주차장정보.csv", encoding='euc-kr')

print(Gangnam_Parking.shape)
Gangnam_Parking.head()
```

```python
Gwanak_Parking = pd.read_csv("관악구주차장정보.csv", encoding='euc-kr')

print(Gwanak_Parking.shape)
Gwanak_Parking.head()
```

```python
Dongdaemun_Parking = pd.read_csv("동대문주차장정보.csv", encoding='euc-kr')

print(Dongdaemun_Parking.shape)
Dongdaemun_Parking.head()
```

### 서울 전체의 1000개의 주차장중 비율

```python
charge = Seoul_Parking[Seoul_Parking["유무료 구분명"] == "유료"]
print(charge.shape)
charge.head()
```

```python
free = Seoul_Parking[Seoul_Parking["유무료 구분명"] == "무료"]
print(free.shape)
free.head()
```

```python
charge_len = len(charge.index)
free_len = len(free.index)
```

```python
free_or_not_seoul = pd.pivot_table(Seoul_Parking, index="유무료 구분명", values="주차장 종류명", 
              aggfunc="count")
```

```python
seoul_pay = pd.pivot_table(charge, index="주차장 종류명", values="기본주차요금", 
              aggfunc="mean")
seoul_pay["기본주차요금"] = seoul_pay["기본주차요금"].round(0)
```

```python
seoul_more_pay_time = pd.pivot_table(Seoul_Parking, index="주차장 종류명", values=["추가단위시간(분)", "추가요금단위"], 
              aggfunc="mean")
seoul_more_pay_time["추가단위시간(분)"] = seoul_more_pay_time["추가단위시간(분)"].round(0)
seoul_more_pay_time["추가요금단위"] = seoul_more_pay_time["추가요금단위"].round(0)
```
